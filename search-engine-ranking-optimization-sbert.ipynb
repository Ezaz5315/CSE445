```python
from sentence_transformers import SentenceTransformer, util

# Load a pretrained Sentence-BERT model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
def rank_documents(query, documents, model):
    query_embedding = model.encode(query, convert_to_tensor=True)
    doc_embeddings = model.encode(documents, convert_to_tensor=True)

    # Compute cosine similarity between query and each document
    scores = util.pytorch_cos_sim(query_embedding, doc_embeddings)[0].cpu().numpy()

    # Rank documents based on similarity scores
    ranked_results = sorted(zip(documents, scores), key=lambda x: x[1], reverse=True)

    return ranked_results
```

```python
# Example query & documents
query = "TensorFlow functions"
documents = [
    "PyTorch is an open-source deep learning framework known for its dynamic computation graph and easy debugging.",
    "TensorFlow is a scalable deep learning library developed by Google, widely used in both research and production.",
    "Keras is a high-level deep learning API built on top of TensorFlow, designed for easy experimentation.",
    "JAX, developed by Google, is a fast and flexible deep learning library optimized for machine learning research.",
    "MXNet is an efficient deep learning framework that supports both symbolic and imperative programming.",
    "Caffe is a deep learning framework popular in computer vision applications due to its speed and efficiency.",
    "Theano is a historical deep learning framework that influenced modern libraries like TensorFlow and PyTorch.",
    "FastAI is a deep learning library built on top of PyTorch, providing simplified high-level APIs.",
    "DL4J (DeepLearning4J) is a deep learning library for Java and Scala, designed for enterprise applications.",
    "PaddlePaddle is a deep learning framework developed by Baidu, optimized for natural language processing and vision."
]

# Get ranked results
ranked_results = rank_documents(query, documents, model)

# Display results
for doc, score in ranked_results:
    print(f"Score: {score:.4f}, Document: {doc}")
```
